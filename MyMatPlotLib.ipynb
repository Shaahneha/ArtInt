# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
z= 2*np.sin(x) ## changing altitude
m = np.sin(2*x) ## changing frequency

# Plot the points using matplotlib
plt.plot(x, y)
#plt.plot(x,z)
plt.plot(x,m)
```

x is simply the range for x axis

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
```
